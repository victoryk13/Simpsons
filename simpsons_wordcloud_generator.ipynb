```python
import pandas as pd
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
df = pd.read_csv("data/simpsons_normalizedtext.csv")
top50_df = pd.read_csv("data/top_50.csv",header=None)
```

```python
df = df.dropna()
```

```python
top50_df
```

```python
top50_df
top50 = {}
            
for row in top50_df.iterrows():
    character = row[1][0]
    temp_df = df.loc[df['raw_character_text']==character]
    merged_text = ""
    for row in temp_df.iterrows():
        current_text = row[1][2]
        merged_text = merged_text + " " + current_text
    
    top50[character] = merged_text
    
```

```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator


for character,text in top50.items():
    
    stopwords = set(STOPWORDS)
    stopwords.add("said")
    stopwords.add("im")
    stopwords.add("ill")
    
    if "simpson" in character.lower():
        mask = np.array(Image.open('masks/'+character.lower()+'.jpg'))
        image_colors = ImageColorGenerator(mask)
        wc = WordCloud(background_color="black", max_words=5000, mask=mask,stopwords=stopwords,
                   min_font_size=6, max_font_size=72 ,relative_scaling=0.9,color_func=image_colors)
    else:
        mask = np.array(Image.open('masks/sillhouette.jpg'))
        wc = WordCloud(background_color="black", max_words=5000,mask=mask,stopwords=stopwords,relative_scaling=0.9)

    # generate word cloud
    wc.generate(text).to_file('wc/'+character.lower()+'_wc.jpg')

    # recolor wordcloud and show
    # we could also give color_func=image_colors directly in the constructor
    # plt.imshow(wc, interpolation="bilinear")
    # plt.axis("off")
    # plt.show()
```
